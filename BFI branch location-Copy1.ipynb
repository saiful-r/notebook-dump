```python
from geopy import ArcGIS, GoogleV3
```

```python
g_arcgis = ArcGIS()

def geocoding(string_address):
    geo_loc = g_arcgis.geocode(string_address)
    return geo_loc.latitude, geo_loc.longitude
```

```python
datacabang_df = datacabang_df[datacabang_df["Alamat"].notnull()]
```

```python
datacabang_df["coor"] = datacabang_df["Alamat"].apply(geocoding)
```

```python
datacabang_df
```

```python
datacabang_df.to_excel("datacabang_coor.xlsx", index=False)
```

```python
import folium
```

```python
datacabang_df[["Lat", "Lon"]] = datacabang_df["coor"].tolist()
```

```python
datacabang_df
```

```python
m = folium.Map(location=[datacabang_df["Lat"].mean(), datacabang_df["Lon"].mean()], zoom_start=5, tiles="OpenStreetMap")

for _, row in datacabang_df.iterrows():
    folium.CircleMarker(
        location=[row["Lat"], row["Lon"]],
        radius=3,
        popup=row[["Cabang", "Branch ID"]],
        color="blue",
        fill=True,
        fill_color="blue"
    ).add_to(m)
    
m
```
